```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_titanic = sns.load_dataset('titanic')
df_titanic
```

```python
df_titanic.iloc[0].age
```

```python
df_titanic.describe()
```

```python
homens = df_titanic[df_titanic['sex'] == 'male'].shape
mulheres = df_titanic[df_titanic['sex'] == 'female'].shape
print('homens:', homens[0], 'mulheres:', mulheres[0], 'Total:', homens[0] + mulheres[0])
```

```python
numero_de_amostra = df_titanic.shape[0]

sobreviventes = df_titanic [df_titanic['alive'] == 'yes'].shape[0]
vitimas = df_titanic[df_titanic['alive'] == 'no'].shape[0]

print('Sobreviventes:', sobreviventes, 'Vitimas:', vitimas, 'Total:', sobreviventes + vitimas)
print('Sobreviventes (%):', (sobreviventes/numero_de_amostra)*100)
print('Vitimas (%):', (vitimas/numero_de_amostra)*100)
```
