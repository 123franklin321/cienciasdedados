```python
import pandas as pd
import plotly.express as px 
import matplotlib.pyplot as plt     
plt.style.use('seaborn')
```

```python
df = pd.read_excel('AdventureWorks.xlsx')
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df.dtypes
```

```python
df['Valor Venda'].sum()
```

```python
df['custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df.head(1)
```

```python
df['custo'].sum()
```

```python
df['lucro'] = df['Valor Venda'] - df['custo']
```

```python
df.head(1)
```

```python
df['lucro'].sum()
```

```python
df['Tempo_Envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(2)
```

```python
df['Tempo_Envio'].dtype
```

```python
df.groupby('Classe')['Tempo_Envio'].mean()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
df.groupby('Marca')['lucro'].sum().sort_values(ascending=False)
```

```python
df.groupby([df['Data Venda'].dt.year, 'Marca'])['lucro'].sum()
```

```python
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False).plot.bar(title='Total produtos vendidos')
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total produtos vendidos");
```

```python
dados = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).reset_index()
```

```python
dados
```

```python
fig = px.bar(dados, x='Quantidade', y='Produto', title='Total Produtos Vendidos', orientation='h')
fig.show()
```

```python
df.groupby(df['Data Venda'].dt.year)['lucro'].sum().plot.bar(title='Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('receita');
```

```python
lucro_ano = df.groupby(df['Data Venda'].dt.year)['lucro'].sum().reset_index()
```

```python
lucro_ano
```

```python
fig = px.bar(lucro_ano, x='Data Venda', y='lucro')
fig.show()
```

```python
df['mes_ano'] = df['Data Venda'].dt.to_period('M')
df.sample(3)
```

```python
lucro_mes = df.groupby(df['mes_ano'])['lucro'].sum().reset_index()
```

```python
lucro_mes.dtypes
```

```python
lucro_mes['mes_ano'] = lucro_mes['mes_ano'].astype(str)
```

```python
lucro_mes
```

```python
fig = px.line(lucro_mes, x='mes_ano', y='lucro')
fig.show()
```

```python
df.groupby('Marca')['lucro'].sum().plot.bar(title='Lucro x Marca');
```

```python
df.groupby('Classe')['lucro'].sum().plot.pie(title='Lucro x Classe');
```

```python
lucro_marca = df.groupby('Classe')['lucro'].sum().reset_index()
```

```python
lucro_marca
```

```python
fig = px.pie(lucro_marca, values='lucro', names='Classe', title='Lucro x Classe')
fig.show()
```
