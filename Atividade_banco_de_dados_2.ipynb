```python
import pandas as pd
import plotly.express as px 
import matplotlib.pyplot as plt     
plt.style.use('seaborn')
import numpy as np
```

```python
df = pd.read_csv('BTC_USD_Price_Prediction_Data.csv')
df
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
dados = df.groupby('24h Open (USD)')['24h High (USD)'].sum().sort_values(ascending=True).reset_index()
```

```python
dados
```

```python
fig = px.bar(dados, x='24h High (USD)', y='24h Open (USD)', orientation='h')
fig.show()
```
