```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
pd.crosstab(df['sex'], df['smoker'], margins=True)
```

```python
pd.crosstab(df['sex'], df['smoker'], normalize='index')
```

```python
pd.crosstab(df['sex'], df['smoker'], normalize='columns')
```

```python
sns.heatmap(
    pd.crosstab(df['sex'], df['smoker']),
    annot=True, cmap='Blues'
);
```

```python
sns.jointplot(data=df, x='total_bill', y='tip');
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex');
```

```python
df.groupby('sex')['tip'].agg(['mean', 'min', 'max', 'median', 'std'])
```

```python
df.groupby(['sex', 'smoker'])['tip'].agg(['mean', 'min', 'max', 'median', 'std'])
```

```python
sns.boxplot(data=df, y='tip', x='sex', hue='smoker');
```
