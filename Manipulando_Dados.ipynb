```python
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders_filtrado = df_orders[(df_orders['order_purchase_timestamp']>=datetime.datetime(2017,11,1)) & (df_orders['order_purchase_timestamp']<= datetime.datetime(2017,11,30))]
df_orders_filtrado
```

```python
pedidos_entregues = df_orders.query('order_status == "delivered"')
duracao_envio_entrega = pedidos_entregues['order_delivered_customer_date'] + pedidos_entregues['order_delivered_carrier_date'] 
duracao_envio_entrega
```

```python
df_orders.set_index('order_purchase_timestamp').resample('M')
```

```python
df_orders.set_index('order_purchase_timestamp').resample('M')['order_id'].count() 
```

```python
df_orders[['order_purchase_timestamp']].drop_duplicates().set_index('order_purchase_timestamp').resample('12H').asfreq()
```

```python
df_orders['ano_mes_compra'] = df_orders['order_purchase_timestamp'].dt.to_period('M')
pedidos_por_mes_ano = df_orders.groupby('ano_mes_compra')[['order_id']].count()
pedidos_por_mes_ano.index = pedidos_por_mes_ano.index.to_timestamp()
pedidos_por_mes_ano
```

```python
new_date_index_range = pd.date_range(start='2016-01-01',end='2018-12-01',freq='MS')
pedidos_por_mes_ano.reindex(new_date_index_range)
```
