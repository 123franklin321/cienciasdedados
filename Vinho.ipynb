```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
link = 'https://raw.githubusercontent.com/smalvar/WineQuality/master/winequality.csv'
df = pd.read_csv(link, sep=';')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.alcohol.unique()
```

```python
for vinho in df['alcohol']:
    try:
      float(vinho)
    except:
        df = df[df.alcohol != vinho]
df['alcohol'] = df['alcohol'].astype(float)
```

```python
df.isnull().sum()
```

```python
wine_type =pd.get_dummies(df['type'])
wine = df.join(wine_type)
wine_type = wine.drop(['type'],axis=1)
```

```python
wine.head()
```

```python
wine.rename(columns= {'fixed acidity': 'Acidez fixa'}, inplace=True)
```

```python
wine.head()
```

```python
import seaborn as sns
```

```python
fig = plt.figure(figsize=(15,15))
axs = []

axs.append(fig.add_subplot(4,3,1))
axs.append(fig.add_subplot(4,3,2))
axs.append(fig.add_subplot(4,3,3))
axs.append(fig.add_subplot(4,3,4))
axs.append(fig.add_subplot(4,3,5))
axs.append(fig.add_subplot(4,3,6))
axs.append(fig.add_subplot(4,3,7))
axs.append(fig.add_subplot(4,3,8))
axs.append(fig.add_subplot(4,3,9))
axs.append(fig.add_subplot(4,3,10))
axs.append(fig.add_subplot(4,3,11))

for col, ax in zip(wine.select_dtypes(include='float64').columns, axs):
    sns.boxplot(x='quality',
                y=col,
                data=wine,
                ax=ax)
plt.show() 
```

```python
df_white = wine[wine['type']== 'White']
df_red = wine[wine['type']== 'Red']

df_white_corr = df_white.corr()
df_red_corr = df_red.corr()
corr = df_red_corr - df_white_corr
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(corr, cmap='cool')
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df_red_corr, cmap='cool')
```

```python
wine['Categoria'] = np.where(wine.quality > 5, 'Regular', 'Ruim')
wine['Categoria'] = np.where(wine.quality > 8, 'Bom', wine.Categoria)
wine.head()
```

```python
wine['Categoria'].value_counts()
```

```python
sns.pairplot(data = wine, x_vars = list(wine.select_dtypes(include='float64').columns[:3]),
             y_vars = list(wine.select_dtypes(include='float64').columns[:3]), hue='Categoria', size=4)
```

```python
sns.pairplot(data = wine, x_vars = list(wine.select_dtypes(include='float64').columns[4:7]),
             y_vars = list(wine.select_dtypes(include='float64').columns[4:7]), hue='Categoria', size=4)
```
