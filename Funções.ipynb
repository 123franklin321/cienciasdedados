```python
df_products = pd.read_csv('olist_products_dataset.csv', usecols=['product_id', 'product_category_name', 'product_photos_qty'])
df_products
```

```python
df_products['rank'] = df_products['product_photos_qty'].rank()
df_products.sort_values(by='rank').dropna()
```

```python
df_products['rank'] = df_products['product_photos_qty'].rank(method='min')
df_products.sort_values(by='rank').dropna()
```

```python
df_products['rank'].value_counts().sort_index()
```

```python
df_products['rank'] = df_products['product_photos_qty'].rank(method='dense')
df_products.sort_values(by='rank').dropna()
```

```python
df_products['rank'] = df_products.groupby(['product_category_name'])[['product_photos_qty']].rank('dense', ascending=False)
df_products.sort_values(by=['product_category_name', 'rank']).dropna()
```
