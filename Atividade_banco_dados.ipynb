```python
df.groupby('High')['Low'].sum().sort_values(ascending=True).plot.bar
plt.show()
```

```python
dados = df.groupby('Volume')['Close'].sum().sort_values(ascending=True).reset_index()
```

```python
dados
```

```python
fig = px.bar(dados, x='Close', y='Volume', orientation='h')
fig.show()
```
