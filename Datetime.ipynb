```python
produto = pd.read_csv('olist_products_dataset.csv', usecols= ['product_id', 'product_category_name'], index_col= 'product_id')
produto
```

```python
pedido = pd.read_csv('olist_order_items_dataset.csv', usecols= ['order_id', 'product_id', 'price', 'order_item_id'], index_col='product_id')
pedido
```

```python
categoria_preco_por_pedido = pd.merge(pedido, produto, on= 'product_id', how= 'inner')
categoria_preco_por_pedido
```

```python
categoria_preco_por_pedido.groupby('product_category_name')['price'].mean().sort_values(ascending=False).head(10)
```

```python
pedido_data = pd.read_csv('olist_orders_dataset.csv', usecols= ['order_id', 'order_status', 'order_purchase_timestamp'], index_col= ['order_id'])
pedido-pedido_data
```

```python
pedido_data_filtrado = pedido_data.query('order_status in ("delivered", "shipped", "invoiced")')
pedido_data_filtrado
```

```python
pedido_data_filtrado.info()
```

```python
pedido_data_filtrado.order_purchase_timestamp
```

```python
def converte_str_para_data(x):
    x = x[:10]
    x = datetime.strptime(x, '%Y-%m-%d')
    return x

pedido_data_filtrado.order_purchase_timestamp.apply(converte_str_para_data)
```

```python
pedido_data_filtrado['data_dt'] = pd.to_datetime(pedido_data_filtrado.order_purchase_timestamp)
pedido_data_filtrado.info()
```

```python
filtro_ano = pedido_data_filtrado.data_dt.dt.year == 2018
pedido_data_filtrado_2018 = pedido_data_filtrado[filtro_ano]
pedido_data_filtrado_2018
```

```python
pd.merge(pedido_data_filtrado_2018, categoria_preco_por_pedido, on= 'order_id', how= 'inner')\
.groupby('product_category_name')['price'].sum().sort_values(ascending=False).head(10)
```

```python
avaliacoes_pedidos = pd.read_csv('olist_order_reviews_dataset.csv', usecols= ['order_id', 'review_score', 'review_comment_title', 'review_comment_message', 'review_creation_date'], index_col=['order_id'])
avaliacoes_pedidos.head()
```

```python
filtro_mais_que_um_produto = pedido.order_item_id > 1
id_dos_pedidos_com_mais_de_um_produto = pedido[filtro_mais_que_um_produto].order_id.unique()
id_dos_pedidos_com_mais_de_um_produto
```

```python
pedidos_com_apenas_um_produto = pedido[~ pedido.order_id.isin(id_dos_pedidos_com_mais_de_um_produto)]
pedidos_com_apenas_um_produto
```

```python
pedidos_com_apenas_um_produto.order_item_id.value_counts()
```

```python
avaliacoes_pedidos_um_produto = pd.merge(pedidos_com_apenas_um_produto, avaliacoes_pedidos, on= 'order_id', how= 'inner')
avaliacoes_pedidos_um_produto.head()
```
