```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
```

```python
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
```

```python
plt.bar([x + 5 for x in histogram.keys()],
    histogram.values(),
    10,
    edgecolor=(0, 0, 0))

plt.axis([-5, 105, 0, 5])

plt.xticks([10 * i for i in range(11)])
plt.xlabel('Notas')
plt.ylabel('Nº de Estudantes')
plt.title('Distribuição das Notas da 1º Avaliação')
plt.show()
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
```

```python
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, bias_squared, 'b:', label='total error')

plt.legend(loc=9)
plt.xlabel('Complexidade do Modelo')
plt.xticks([])
plt.title('O Dilema Viés-Variância')
plt.show()
```

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count),
                 xytext=(5, -5),
                 textcoords='offset points')

plt.title('Minutos Diários x Número de Amigos')
plt.xlabel('Nº de Amigos')
plt.ylabel('Minutos diários dedicaados as site')
plt.show()
```
