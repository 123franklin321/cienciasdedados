```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
```

```python
df_products = pd.read_csv('olist_products_dataset.csv')
df_products
```

```python
tb_telefonia = df_products.query('product_category_name == "telefonia" and product_photos_qty > 5')
tb_telefonia
```

```python
tb_eletronicos = df_products.query('product_category_name == "eletronicos" and product_photos_qty > 5')[['product_id', 'product_category_name', 'product_photos_qty']]
tb_eletronicos
```

```python
tb_telefonia.append(tb_eletronicos)
```

```python
pd.concat([tb_telefonia, tb_eletronicos])
```

```python
df_orders = pd.read_csv('olist_orders_dataset.csv', usecols= ['order_id', 'order_status'])
df_orders
```

```python
df_payments = pd.read_csv('olist_order_payments_dataset.csv', usecols= ['order_id', 'payment_type', 'payment_value'])
df_payments
```

```python
orders = df_orders.set_index('order_id').sort_index()
```

```python
payments = df_payments.groupby('order_id')[['payment_value']].sum()
payments
```

```python
pd.concat([orders, payments], axis=1)
```

```python
df_orders.merge(df_payments, on= 'order_id')
```

```python
df_items = pd.read_csv('olist_order_items_dataset.csv')
df_items
```

```python
df_reviews = pd.read_csv('olist_order_reviews_dataset.csv')
df_reviews
```

```python
print(df_items.info(), df_reviews.info())
```

```python
df_items.merge(df_reviews.dropna(subset= ['review_comment_message']), on= 'order_id', how= 'right').sort_values(by=['order_id', 'product_id'])
```
